```python
import numpy as np
import pandas as pd
```

```python
import boto3

_BUCKET_NAME = 'shipping-ml'

s3 = boto3.resource('s3')
my_bucket = s3.Bucket(_BUCKET_NAME)

lista_archivos = []
for object_summary in my_bucket.objects.filter(Prefix="data/raw/daily/raw/DataGuido/train_images/"):
    lista_archivos.append(object_summary.key)
                
```

```python
s3 = boto3.resource('s3')
my_bucket = s3.Bucket(_BUCKET_NAME)

lista_archivos_val = []
for object_summary in my_bucket.objects.filter(Prefix="data/raw/daily/raw/DataGuido/train_label/"):
    lista_archivos_val.append(object_summary.key)
```

```python
np.asarray(lista_archivos).shape, np.asarray(lista_archivos_val).shape, np.asarray(lista_archivos).shape[0] - np.asarray(lista_archivos_val).shape[0] 
```

```python
file_train_name_label = []
for file in lista_archivos_lab:
    file_train_name_label.append(file.split('/')[-1].replace('msk_patch', 'img_patch'))
```

```python
file_train_name_label = pd.DataFrame(file_train_name_label)
```

```python
file_train_name_image = []
for file in lista_archivos:
    file_train_name_image.append(file.split('/')[-1])
```

```python
file_train_name_image = pd.DataFrame(file_train_name_image).rename(columns={0:'name_image'})
file_train_name_label = pd.DataFrame(file_train_name_label).rename(columns={0:'name_label'})
```

```python
file_train_name_image.shape, file_train_name_label.shape
```

```python
both = pd.merge(file_train_name_label, file_train_name_image, left_on='name_label', right_on='name_image', how='left')
```
